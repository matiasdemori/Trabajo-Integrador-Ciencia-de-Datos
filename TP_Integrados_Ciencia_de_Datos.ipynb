##Preparación de los datos

Realizamos un recorte temporal del dataset, tomando como referencia abril de 2026 y seleccionando los registros correspondientes a un año hacia atrás.

Este filtro fue la segmentacion planteada para el recorte de datos, haciendo que el análisis sea más manejable y esté alineado con el alcance definido para el trabajo. Además, nos permite comparar los precios de combustibles dentro de un intervalo temporal concreto, evitando mezclar datos históricos demasiado antiguos con datos actuales.

```python
# Nos aseguramos de que la columna de fecha esté en formato datetime
df["fecha_vigencia"] = pd.to_datetime(df["fecha_vigencia"], errors="coerce")

# Definimos el rango de fechas que queremos analizar
fecha_inicio = "2025-04-01"
fecha_fin = "2026-04-30"

# Filtramos el dataset para quedarnos solo con los registros entre abril 2025 y abril 2026
df_recorte = df[
    (df["fecha_vigencia"] >= fecha_inicio) &
    (df["fecha_vigencia"] <= fecha_fin)
].copy()

# Verificamos el tamaño del nuevo dataset filtrado
df_recorte.shape
```

```python
df_recorte["fecha_vigencia"].min(), df_recorte["fecha_vigencia"].max()
```

Con este código agrupamos los registros del dataset recortado por mes.

Esto nos sirve para analizar si la cantidad de datos está distribuida de manera relativamente pareja entre abril de 2025 y abril de 2026, o si existen meses con mayor o menor cantidad de registros.

Buscamos posibles valores atípicos en la variable precio utilizando el método del rango intercuartílico.

Esto nos sirve para detectar precios que se encuentran muy alejados del comportamiento general de los datos. Estos casos pueden representar errores de carga, situaciones excepcionales o diferencias reales entre productos, empresas o provincias.

```python
# Calculamos los cuartiles y el rango intercuartílico
Q1 = df_limpio["precio"].quantile(0.25)
Q3 = df_limpio["precio"].quantile(0.75)
IQR = Q3 - Q1

# Definimos los límites para detectar valores atípicos
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

print("Límite inferior:", limite_inferior)
print("Límite superior:", limite_superior)

# Filtramos los registros considerados valores atípicos
outliers_precio = df_limpio[
    (df_limpio["precio"] < limite_inferior) |
    (df_limpio["precio"] > limite_superior)
]

outliers_precio
```

###Valores atípicos

Calculamos cuántos registros fueron identificados como valores atípicos y qué porcentaje representan sobre el total del dataset.

Esto nos ayuda a evaluar el impacto potencial de estos valores en el análisis. Si el porcentaje es bajo, podrían analizarse como casos particulares; si es alto, habría que revisar si responden a diferencias reales del mercado o a problemas de calidad de datos.

```python
cantidad_outliers = outliers_precio.shape[0]
porcentaje_outliers = cantidad_outliers / df_limpio.shape[0] * 100

print("Cantidad de valores atípicos:", cantidad_outliers)
print("Porcentaje de valores atípicos:", porcentaje_outliers)
```

```python
precio_por_provincia = (
    df_limpio
    .groupby("provincia")["precio"]
    .mean()
    .sort_values(ascending=False)
)

precio_por_provincia
```

Agrupamos los datos por empresa o bandera comercial y calculamos el precio promedio de cada una.

Esto nos permite observar si existen diferencias de precios entre empresas comercializadoras y analizar si algunas banderas presentan precios sistemáticamente más altos o más bajos.

```python
precio_por_empresa = (
    df_limpio
    .groupby("empresabandera")["precio"]
    .mean()
    .sort_values(ascending=False)
)

precio_por_empresa
```

Agrupamos los datos por mes y calculamos el precio promedio mensual.

Esto nos permite observar la evolución temporal de los precios durante el período analizado y detectar si existe una tendencia creciente, decreciente o estable a lo largo del tiempo.

```python
# Creamos una columna con año y mes
df_limpio["anio_mes"] = df_limpio["fecha_vigencia"].dt.to_period("M")

# Calculamos el precio promedio por mes
precio_por_mes = (
    df_limpio
    .groupby("anio_mes")["precio"]
    .mean()
)

precio_por_mes
```

Calculamos el precio promedio combinando provincia y tipo de producto.

Esto nos permite observar si las diferencias de precios entre combustibles se mantienen en todas las provincias o si existen provincias donde determinados productos presentan valores más altos o más bajos que el promedio general.

```python
precio_producto_provincia = (
    df_limpio
    .groupby(["provincia", "producto"])["precio"]
    .mean()
    .reset_index()
    .sort_values("precio", ascending=False)
)

precio_producto_provincia
```

Calculamos el precio promedio combinando empresa comercializadora y tipo de producto.

Esto nos permite analizar si las diferencias de precios entre empresas se mantienen para todos los combustibles o si algunas empresas presentan precios más altos en productos específicos.

```python
precio_empresa_producto = (
    df_limpio
    .groupby(["empresabandera", "producto"])["precio"]
    .mean()
    .reset_index()
    .sort_values("precio", ascending=False)
)

precio_empresa_producto
```

Para explorar si existe una relación entre el tiempo y el precio, transformamos los meses en una variable numérica y calculamos su correlación con el precio promedio mensual.

Esto nos permite tener una primera aproximación sobre si los precios tienden a aumentar o disminuir durante el período analizado.

```python
# Convertimos el precio promedio mensual en un DataFrame
precio_por_mes_df = precio_por_mes.reset_index()

# Creamos una variable numérica para representar el orden de los meses
precio_por_mes_df["mes_numero"] = range(1, len(precio_por_mes_df) + 1)

# Calculamos la correlación entre el paso del tiempo y el precio promedio
correlacion_tiempo_precio = precio_por_mes_df["mes_numero"].corr(precio_por_mes_df["precio"])

correlacion_tiempo_precio
```

Calculamos el precio promedio y el desvío estándar por empresa comercializadora. Esto nos permite identificar no solo qué empresas tienen precios más altos, sino también cuáles presentan mayor variabilidad en sus precios.

```python
dispersion_por_empresa = (
    df_limpio
    .groupby("empresabandera")["precio"]
    .agg(["mean", "std", "min", "max", "count"])
    .sort_values("std", ascending=False)
)

dispersion_por_empresa
```

Calculamos el precio promedio y el desvío estándar por provincia. Esto nos permite observar en qué provincias los precios son más variables y en cuáles se comportan de manera más homogénea.

```python
dispersion_por_provincia = (
    df_limpio
    .groupby("provincia")["precio"]
    .agg(["mean", "std", "min", "max", "count"])
    .sort_values("std", ascending=False)
)

dispersion_por_provincia
```

Calculamos la diferencia mensual entre el producto con mayor precio promedio y el producto con menor precio promedio. Esto nos permite analizar si la brecha entre combustibles se mantiene estable o si cambia durante el período estudiado.

```python
# Calculamos el precio promedio mensual por producto
precio_mensual_producto = (
    df_limpio
    .groupby(["anio_mes", "producto"])["precio"]
    .mean()
    .reset_index()
)

# Pasamos la tabla a formato comparativo
tabla_mensual_producto = precio_mensual_producto.pivot(
    index="anio_mes",
    columns="producto",
    values="precio"
)

# Calculamos la brecha mensual entre el producto más caro y el más barato
tabla_mensual_producto["brecha_mensual"] = (
    tabla_mensual_producto.max(axis=1) - tabla_mensual_producto.min(axis=1)
)

tabla_mensual_producto
```
