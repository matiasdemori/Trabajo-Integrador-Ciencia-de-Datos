###3.1.A Business Understanding (Comprensión del negocio)


<u>_**Contexto**_</u>


El precio de los combustibles tiene un impacto directo en la economía argentina, ya que influye en transporte, logística,
costos operativos de empresas y precio final de bienes y servicios.
Además, las variaciones de precios afectan tanto a consumidores individuales como a actividades productivas en todo el país.
Debido a esto, resulta relevante estudiar cómo evolucionan los precios y si existen diferencias entre regiones, ya que esa información puede aportar una mejor comprensión del mercado de combustibles argentino.


<u>_**Problema / Necesidad**_</u>


Actualmente existe una gran cantidad de información pública sobre precios de combustibles, pero al estar distribuida en múltiples registros históricos y actualizaciones frecuentes,
no siempre resulta sencillo interpretar tendencias recientes ni comparar provincias de forma clara.
En este proyecto se busca transformar esos datos en información útil, permitiendo observar la evolución reciente de precios y detectar diferencias geográficas relevantes.


<u>_**Objetivo principal**_</u>

- Analizar la evolución de los precios promedio semanales de combustibles en Argentina durante los últimos 12 meses disponibles.

<u>_**Objetivos secundarios**_</u>

- Comparar precios promedio entre provincias.

- Detectar provincias con valores sistemáticamente más altos o más bajos.

- Identificar períodos con aumentos relevantes.

- Observar patrones similares o diferentes entre regiones.

- Comunicar resultados mediante visualizaciones claras.

<u>_**Pregunta analítica principal**_</u>

- ¿Cómo evolucionaron los precios promedio semanales de los combustibles en Argentina durante los últimos 12 meses?

<u>_**Preguntas analíticas secundarias**_</u>

- ¿Qué provincias presentaron los precios promedio más altos y más bajos?
- ¿Hubo semanas con aumentos más pronunciados que otras?
- ¿Existen comportamientos regionales similares en la evolución de precios?

<u>_**Criterios de éxito**_</u>

Se considerará exitoso el proyecto si se logra:

- Construir una serie temporal clara de precios promedio.

- Detectar tendencias relevantes en el período analizado.

- Comparar provincias de manera objetiva.

- Presentar conclusiones respaldadas por datos y gráficos adecuados.
